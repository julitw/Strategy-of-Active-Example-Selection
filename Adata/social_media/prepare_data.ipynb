```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('train.tsv', sep='\t')
df.dropna(axis=0, inplace=True)
```

```python
df = df[['Label', 'Text_data']]
```

```python
df['label'] = df['Label']
df['text'] = df['Text_data']
```

```python
df
```

```python
np.unique(df['label'])
```

```python
df
```

```python
df
```

```python
train_df, test_df = train_test_split(df, test_size=0.12, random_state=42, stratify=df['label'])
```

```python
test_df
```

```python
test_df['label'].value_counts()
```

```python

train_sampled = (
   train_df.groupby('label', group_keys=False)
    .apply(lambda x: x.sample(min(15, len(x)), random_state=42))
    .sample(frac=1, random_state=42)  
    .reset_index(drop=True)
)

train_sampled2 = (
   train_df.groupby('label', group_keys=False)
    .apply(lambda x: x.sample(min(15, len(x)), random_state=43))
    .sample(frac=1, random_state=43)  
    .reset_index(drop=True)
)

train_sampled3 = (
   train_df.groupby('label', group_keys=False)
    .apply(lambda x: x.sample(min(15, len(x)), random_state=44))
    .sample(frac=1, random_state=442)  
    .reset_index(drop=True)
)

train_sampled4 = (
   train_df.groupby('label', group_keys=False)
    .apply(lambda x: x.sample(min(15, len(x)), random_state=45))
    .sample(frac=1, random_state=45)  
    .reset_index(drop=True)
)

```

```python
train_sampled[['text', 'label']].to_csv('../../examples_for_prompt/social_media/cohere/random.csv', index=False)
train_sampled[['text', 'label']].to_csv('../../examples_for_prompt/social_media/gpt4omini/random.csv', index=False)
train_sampled[['text', 'label']].to_csv('../../examples_for_prompt/social_media/llama3/random.csv', index=False)

train_sampled2[['text', 'label']].to_csv('../../examples_for_prompt/social_media/cohere/random2.csv', index=False)
train_sampled2[['text', 'label']].to_csv('../../examples_for_prompt/social_media/gpt4omini/random2csv', index=False)
train_sampled2[['text', 'label']].to_csv('../../examples_for_prompt/social_media/llama3/random2.csv', index=False)


train_sampled3[['text', 'label']].to_csv('../../examples_for_prompt/social_media/cohere/random3.csv', index=False)
train_sampled3[['text', 'label']].to_csv('../../examples_for_prompt/social_media/gpt4omini/random3.csv', index=False)
train_sampled3[['text', 'label']].to_csv('../../examples_for_prompt/social_media/llama3/random3.csv', index=False)


train_sampled4[['text', 'label']].to_csv('../../examples_for_prompt/social_media/cohere/random4.csv', index=False)
train_sampled4[['text', 'label']].to_csv('../../examples_for_prompt/social_media/gpt4omini/random4.csv', index=False)
train_sampled4[['text', 'label']].to_csv('../../examples_for_prompt/social_media/llama3/random4.csv', index=False)

```
