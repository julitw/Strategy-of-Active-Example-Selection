```python
from datasets import load_dataset
import pandas as pd

# SST-5 (default = 5 klas)
dataset = load_dataset("sst", "default")
```

```python

train_dataset = dataset['train'].to_pandas()
```

```python
print(train_dataset['label'].value_counts())
```

```python
train_dataset
```

```python

train_dataset['label_binned'] = pd.cut(train_dataset['label'], bins=5, labels=[1, 2,3 , 4, 5])
print(train_dataset['label_binned'].value_counts())
```

```python
train_dataset['label_binned'].value_counts()
```

```python
train_subset = train_dataset.groupby('label_binned').apply(lambda x: x.sample(200, random_state=42)).reset_index(drop=True)
train_subset = train_subset.sample(frac=1, random_state=42)  
```

```python
train_subset
```

```python
train_subset = train_subset[['sentence', 'label_binned']]
train_subset.columns = ['text', 'label']
train_subset
```

```python
train_subset.to_csv('train_subset.csv', index=False)
```
