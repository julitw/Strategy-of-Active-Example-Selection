```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from utils import evaluate_predictions_from_folders
import pandas as pd
import re
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
model_name = 'cohere'
folder_paths = [f'results/random_samples_temp0/{model_name}']
```

```python
df = evaluate_predictions_from_folders(folder_paths, True, 'original_label', 'output' )
df.sort_values(by='file')
```

```python
df.to_csv('acc_f1_precision_recall.csv', index=False)
```

```python
# Ekstrahowanie datasetu i liczby próbek (np. random24_40, random42_80 itp.)
df["dataset"] = df["file"].apply(lambda x: x.split(f"_{model_name}_")[0])
df["sample_group"] = df["file"].apply(lambda x: re.search(r'random\d+_\d+', x).group())

# Podział: w każdej grupie datasetu wybierz df z mniejszym sample_group i df z większym
smaller_sample_df = df.groupby("dataset").apply(lambda g: g.loc[g["sample_group"].str.extract(r'_(\d+)').astype(int)[0].idxmin()])
larger_sample_df = df.groupby("dataset").apply(lambda g: g.loc[g["sample_group"].str.extract(r'_(\d+)').astype(int)[0].idxmax()])

# Reset indeksów
smaller_sample_df.reset_index(drop=True, inplace=True)
larger_sample_df.reset_index(drop=True, inplace=True)
```

```python
df["random_id"] = df["sample_group"].apply(lambda x: x.split("_")[0])
df["sample_size"] = df["sample_group"].apply(lambda x: int(x.split("_")[1]))

# Dla każdego datasetu i random_id zostawiamy tylko te grupy, które mają dokładnie dwie różne liczności próbek (np. 40 i 80)
valid_groups = (
    df.groupby(["dataset", "random_id"])["sample_size"]
    .nunique()
    .reset_index()
    .query("sample_size == 2")
    [["dataset", "random_id"]]
)

# Łączymy z oryginalnym dataframe, aby zostawić tylko odpowiednie pary
filtered_df = df.merge(valid_groups, on=["dataset", "random_id"])

filtered_df
```

```python
# Grupujemy po dataset i sample_size, obliczamy średnią i odchylenie standardowe dla metryk
metrics = ["accuracy", "f1_score", "recall", "precision"]
agg_df = (
    df.groupby(["dataset", "sample_size"])[metrics]
    .agg(["mean", "std"])
    .reset_index()
)

# Dzielimy na dwa: mniejsze i większe sample_size w ramach każdego datasetu
min_df = agg_df.groupby("dataset").apply(lambda g: g.loc[g["sample_size"].idxmin()])
max_df = agg_df.groupby("dataset").apply(lambda g: g.loc[g["sample_size"].idxmax()])

# Reset indeksów
min_df.reset_index(drop=True, inplace=True)
max_df.reset_index(drop=True, inplace=True)

```

```python
# Spłaszczanie kolumn MultiIndex
min_df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in min_df.columns.values]
max_df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in max_df.columns.values]

# Przemnożenie i zaokrąglenie metryk
for metric in ['accuracy', 'f1_score', 'recall', 'precision']:
    min_df[f'{metric}_mean'] = (min_df[f'{metric}_mean'] * 100).round(1)
    min_df[f'{metric}_std'] = (min_df[f'{metric}_std'] * 100).round(1)
    max_df[f'{metric}_mean'] = (max_df[f'{metric}_mean'] * 100).round(1)
    max_df[f'{metric}_std'] = (max_df[f'{metric}_std'] * 100).round(1)

```
