```python
import pandas as pd
import random
```

```python
dataset_tag = "yahoo"
seed = 41
dictt =  yahoo
n = dictt['num_samples']
total = dictt['num_samples'] * dictt['num_classes']
```

```python
random.seed(seed)
```

```python
df = pd.read_csv(f"../Adata/{dataset_tag}/train_subset.csv")
df
```

```python
df_sample = df.groupby('label').sample(n=n, random_state=seed)
df_sample
```

```python
df_shuffled = df_sample.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled
```

```python
seed = 42
df_shuffled.to_csv(f'selected_samples/{dataset_tag}/gpt4omini_random{seed}_{total}.csv', index=False)
df_shuffled.to_csv(f'selected_samples/{dataset_tag}/llama3_random{seed}_{total}.csv', index=False)
df_shuffled.to_csv(f'selected_samples/{dataset_tag}/cohere_random{seed}_{total}.csv', index=False)
```
